```python
from entsoe import EntsoePandasClient
from entsoe.mappings import BIDDING_ZONES, TIMEZONE_MAPPINGS
from entsoe.parsers import parse_loads
from bentso import TOKEN


class BetterENTSOE(EntsoePandasClient):
    def query_hydro_filling_rate(self, country_code, start, end):
        """
        Parameters
        ----------
        country_code : str
        start : pd.Timestamp
        end : pd.Timestamp
        Returns
        -------
        str
        """
        domain = BIDDING_ZONES[country_code]
        params = {
            'documentType': 'A72',
            'processType': 'A16',
            'in_Domain': domain
        }
        response = self.base_request(params=params, start=start, end=end)
        series = parse_loads(response.text)
        series = series.tz_convert(TIMEZONE_MAPPINGS[country_code])
        return series
```

```python
b_client = BetterENTSOE(TOKEN)
```
